### 2.3.5 - Tipo de dados str

O tipo de dados str possui a capacidade de armazenar uma sequência de caracteres. Essa sequência pode ser armazenada com aspas simples ou duplas, desde que haja consistência em sua utilização, ou seja, se você iniciar uma sequência com aspas simples, deve fechá-la também com aspas simples, utilizando o mesmo preceito para as aspas duplas.

Veja alguns exemplos de declaração de variáveis do tipo str a seguir:

```python
nome = 'Fulano'
sobrenome = ' de Tal'
nome_completo = nome + sobrenome + ' da Silva'
nome_completo
```

A linguagem Python permite a criação de textos dispostos em mais de uma linha. Para isso, você deve informar uma sequência de três aspas, simples ou duplas, conforme o exemplo a seguir. As aspas simples ou duplas poderão, então, ser inseridas no texto normalmente.

```python
texto_longo = '''
Primeira linha do texto!
Segunda linha do texto!
Terceira linha do texto! '''

texto_longo
print(texto_longo)
```

## 2.4 - Leitura de variáveis

Você pode realizar a leitura de variáveis pela instrução input, que permite (opcionalmente) informar uma mensagem na tela, conforme o exemplo a seguir.

```python
nome = input("Digite o seu nome: ")
print(nome)
```

Para ler um número é preciso fazer a conversão, pois, por padrão, a instrução **input** realiza a leitura de uma variável do tipo str.

```python
num_inteiro = int(input("Digite um numero inteiro: "))
print(num_inteiro)
```

## 2.5 - Conversão de tipos de dados


O Python realiza automaticamente a conversão do tipo de dados de uma variável quando um valor é atribuído a ela. Nos exemplos a seguir, a variável n recebe valores dos tipos inteiro, real e texto:

```python
n = 6
m = 6.6
o = "seis"
print(m,","n,","o)
```

Entre variáveis numéricas não é preciso forçar a conversão de dados para realizar operações, uma vez que a conversão ocorre de modo implícito.

```python
n1 = 2
n2 = 7.67
n1 = n2 + n1 
print(n1)
```

No entanto, a conversão de número para texto precisa ser explícita, e vice-versa. Os exemplos a seguir apresentam a conversão de número para texto e de texto para número, respectivamente:

```python
num = input("35")
n = str(38.78)
t = float('789.85')
print(num, n, t)
```
